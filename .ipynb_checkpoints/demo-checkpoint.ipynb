# Clustering Text 

In the following notebook, we try to extract sense from text data using clustering. 

```python
from utils import import_data, preprocess
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
import re
from vect_utils import c_distance, simple_tokenizer
from scipy.sparse import csr_matrix
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from cluster import K_Means
import numpy as np

```

```python



filepath = 'data.txt'

data_as_lines, data_as_words = import_data(filepath)
l = preprocess(data_as_words)


vectorizer = TfidfVectorizer(
    use_idf=True, tokenizer=simple_tokenizer, 
    max_features=100,
    stop_words='english')

X = vectorizer.fit_transform(l)

from sklearn.decomposition import PCA
pca = PCA(n_components=50)
reduced = pca.fit_transform(X.todense())



from itertools import cycle
cycol = cycle('bgrcmk')


t = reduced.transpose()
print len(t[0])
for k in xrange(t.shape[1]):
    plt.scatter(t[3], t[4], c=np.random.rand(3,))
plt.xlim(-0.04, -0.01)
plt.ylim(-0.1, 0.05)
plt.show()
# fig = plt.figure()
# ax = fig.add_subplot(111, projection='3d')
# 
#     ax.scatter(t[0][k], t[1][k], t[2][k], c=cycol.next())
# plt.show()

```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
for k in xrange(t.shape[1]):
    ax.scatter(t[0][k], t[1][k], t[2][k], c=cycol.next())
plt.xlim(-0.04, -0.01)
plt.ylim(-0.1, 0.05)
plt.show()
```

```python
number_of_clusters = 5
km = KMeans(n_clusters=number_of_clusters)

km_own = K_Means(k=5)
km_own.fit(X.todense())
print km_own.get_clusters()

```

```python
km.fit(X)
order_centroids = km.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

centroids = []
for i in range(number_of_clusters):
    top_words = [terms[ind] for ind in order_centroids[i, :7]]
    centroids.append(' '.join(top_words))
    print "Cluster {}: {}".format(i, ' '.join(top_words)) 
```
